3) The assumption we make about the grades data is that the file is formatted consistently with headers on the first line, weights on the second line, and student data starting from the third line down. For weights, integers are expected, and string is expected for student data [0] while integers are expected for student data [1:].

```python
def readGrades (filename):
    '''
    Reads and processes a grades file to extract headers, quiz weights, and student grades.

    This function reads a file with the given filename that is specifically formatted for grades. It assumes the file contains:
    - Headers on the first line (e.g., student names and quiz names),
    - Weights for each quiz on the second line,
    - Subsequent lines with student names followed by their quiz scores.

    Parameters:
        filename (str): The path to the file to be read.

    Returns:
        tuple: A tuple containing three elements:
            - headers (list of str): A list of headers, of quiz titles.
            - weights (list of int): A list of integers representing the weights of each quiz.
            - students (list of list): A list of lists where each sublist represents a student and 
            their grades for each quiz, formatted as [student_name, quiz1_grade, quiz2_grade, ...].
    
    '''
    # Helper function to safely convert a string to an integer
    def safeInt(value):
        try:
            return int(value)
        except ValueError: # Returns 0 if conversion fails
            return 0
            
    # open file & read all lines
    with open(filename, 'r') as gradeFile:
        lines = gradeFile.readlines()
        
    # Remove any leading/trailing whitespace from each line
    lines = [line.strip() for line in lines]
    
    # Extract the headers and data
    headers = lines[0].split('\t')[1:]

    
    # Extract weights from the second line
    weight_data = lines[1].split('\t')[2:]    # Skip the first element which is 'Weight'
    weights = [safeInt(value) for value in weight_data]    # Convert weight values to integers using safe_int


    students = []
    for line in lines[2:]: 
        parts = line.split('\t')    # Split the line into parts by tab character
        name = parts[0]    # The first part is the student's name
        scores = [safeInt(score) for score in parts[2:]]    # Convert the remaining parts to integers (scores) using safe_int
        students.append([name] + scores)    # Append the student's name and scores as a list to the students list

    # Return the parsed headers, weights, and students data
    return headers, weights, students
```

```python
def gradeEvaluator(total_score):
    '''
    Evaluates the letter grade based on the total weighted score of a student.

    Parameters:
    - total_score (float): The total weighted score for a student, typically between 0 and 100.

    Returns:
    - str: A string representing the letter grade ("A", "B", "C", "D", or "F").
    '''
    
    # Check if the score is 90 or above, assign grade "A"
    if total_score >= 90:
        return "A"
    # Check if the score is between 80 and 89.99, assign grade "B"
    elif total_score >= 80:
        return "B"
    # Check if the score is between 70 and 79.99, assign grade "C"
    elif total_score >= 70:
        return "C"
    # Check if the score is between 60 and 69.99, assign grade "D"
    elif total_score >= 60:
        return "D"
    # If the score is below 60, assign grade "F"
    else:
        return "F"
```

```python
def computeGrades(quizes, weights, students):
    '''
    Computes the total weighted grades for students based on their quiz scores
    and the associated weights for each quiz.

    Parameters:
    - quizes (list): A list of strings representing the names of the quizzes. 
                     The first entry is assumed to be a placeholder (e.g., "Name") and is skipped.
    - weights (list): A list of floats/integers representing the weight of each quiz.
                      The first entry is assumed to be a placeholder and is skipped.
    - students (list of lists): A list where each sublist contains a student's name and their quiz scores.

    For each student, the function calculates the weighted score for each quiz, sums them up,
    and prints the total score along with the corresponding grade.
    
    Returns:
    None. Prints each student's name, their scores for each quiz, and their final weighted grade.
    '''
    
    
    # Loop through each student and calculate their total weighted score
    for student in students:
        total_score = 0  # Initialize the total score for the student
        
        # Print the student's name
        print(f"Student: \t{student[0]}")
        
        # Calculate and print the weighted score for each quiz
        for quiz, score, weight in zip(quizes, student[1:], weights):
            # Calculate the weighted score for the current quiz
            final_score = score * (weight / 100)
            
            # Accumulate the total weighted score for the student
            total_score += final_score
            
            # Print the quiz details: name, score, and weight
            print(f"{quiz}\t\t{score}\t{weight:.2f}")
        
        # Print the total weighted score and the corresponding grade
        print(f"Total: {total_score:.2f}\t{gradeEvaluator(total_score)}\n")
```

```python
def computeGrades2(quiz_names, weights, students):
    """
    Computes final grades for students by calculating weighted quiz scores
    and dropping the quiz with the smallest impact on the total score.

    Parameters:
    - quiz_names (list): A list of strings representing the names of the quizzes.
    - weights (list): A list of integers/floats representing the weight of each quiz.
    - students (list of lists): A list where each sublist contains a student's name and their quiz scores.

    Each student's lowest-weighted quiz will be dropped from their grade calculation. 
    The remaining quizzes' weights will be normalized to 100% before calculating the final grade.
    
    Returns:
    None. Prints each student's name, their scores, which quiz was dropped, and their final score.
    """
    
    for student in students:
        student_name = student[0]
        student_scores = student[1:]

        print(f"Student: \t{student_name}")
        
        # Initialize lists to track weighted scores and find quiz to drop
        weighted_scores = []
        score_drops = [(score * (weight / 100)) - weight for score, weight in zip(student_scores, weights)]

        # Find the index of the quiz to drop (smallest loss)
        quiz_to_drop_index = score_drops.index(min(score_drops))

        # Print quiz results and mark dropped quiz
        for i, (quiz_name, score, weight) in enumerate(zip(quiz_names, student_scores, weights)):
            drop_status = "drop" if quiz_to_drop_index == i else ""
            print(f"{quiz_name}\t{score}\t{weight}\t{drop_status}")

        # Remove the dropped quiz from scores and weights
        adjusted_scores = [score for i, score in enumerate(student_scores) if i != quiz_to_drop_index]
        adjusted_weights = [weight for i, weight in enumerate(weights) if i != quiz_to_drop_index]

        # Normalize weights to 100%
        total_weight = sum(adjusted_weights)
        normalized_weights = [(100 / total_weight) * weight for weight in adjusted_weights]

        # Calculate the weighted score
        total_score = sum(score * (weight / 100) for score, weight in zip(adjusted_scores, normalized_weights))

        # Print the total score and grade
        print(f"Total: {total_score:.2f}\t{gradeEvaluator(total_score)} \n")
```
