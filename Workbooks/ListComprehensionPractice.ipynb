### Exercise 1:  
Use a list comprehension to generate a list of strings where each string is a lower case letter.

### Exercise 2: 
Design a list comprehension to build a list of all odd numbers between 1 and 99 inclusive.  

```python
oddList = [ num for num in range (1, 99 + 1) if num % 2 != 0 ]  # add list comprehension here
print(oddList)
```

### Exercise 3: 
Design a list comprehension to parse a pair of list of numbers, selecting the smallest value out of the two lists at each index.  Note this is a repeat of the in-class exersise we completed, except you use a list comprehension.

```python
A = [1,2,3,4,-5,0]
B = [3,3,3,3,3]
C = [ a if a < b else b for (a, b) in zip(A, B)]
  # add your list comprehension here
print(C)
# expected output is [1,2,3,3,-5]  
```

### Exercise 4: 
Build a list comprehension that again creates a list of 26 letters, excluding the vowels.  This is a list of consonents. 

```python
consonents = [ c for c in string.ascii_lowercase if c not in ('a' , 'e' , 'i' , 'o', 'u' ) ] # add your list comprehension here
print(consonents)
```

### Exercise 5:
Use a list comprehension to build a list of people who are giving a gift to Martha.  
Note that you have access to a dictionary named *presents* where  
- the key is a person's name giving gifts
- the value is a list of people who are receiving gifts from that person

For example, in the dictionary below James is giving gifts to Sam, Martha and Son.   Thus James should be included in the list you generate.   
  

```python
presents = {'James':['Sam','Martha','Son'], 'Martha':['Son','Cindy'],'Cindy':['Sam','Cassie'],'Beatrice':['Martha','Cindy']}
print(presents)
martha_givers = [ name for name, gifts in presents.items() if 'Martha' in gifts ]
print(martha_givers)
```

### Exercise 6: 
Write a list comprehension that creates a list of tuples (a,b) where a+b = 30 and a and b are both greater than 0.   For example, (10,20) will be in the list because 10+20=30.   But (35,-5) is not allowed because -5 is not greater than 0.  Note that it is ok to put both (10,20) and also (20,10) because tuples are considered to be ordered objects.   

```python
sum_30 = [ (a , b) for a in range (1, 31) for b in range (1, 31) if a + b == 30] # add your list comprehension here
print(sum_30)
```
