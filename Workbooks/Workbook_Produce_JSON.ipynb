### Exercise 1: Open and Read Data
- open the `produce.json` file and read it in using the json.load() function.
- print the data both as a native python structure and also in the "pretty" format using the json.dumps() function.


```python
filename = 'produce.json'
file = open(filename, 'r') 
produce_data = json.load(file)
file.close()
print(json.dumps(produce_data, indent =2))
```

### Exercise 2: Yellow Things
Traverse the python structure and print a list of the names of all things that are yellow in color.

```python
# write a command that will print all yellow things
# Get a list of names for all yellow things
yellow_food = []
for food in produce_data["produce"]:
    if food.get("color") == "yellow":
        yellow_food.append(food["name"])

print(yellow_food) 
```

### Exercise 3: Tasty Things
Again traverse the data and print a list of names of all things that are marked as tasty.  

```python
# write a command that will print all things that I think are tasty
tasty_food_yummy = []
for food in produce_data["produce"]:
    if food.get("tasty") == True:
        tasty_food_yummy.append(food["name"])

print(tasty_food_yummy)
```

### Exercise 4: Fruit vs Veggie
Again traverse the data and keep track of the count of the number of fruits and veggies in 
the data.  Make a brief table showing how many of each are present.  You should have something like  
`
fruit     : 4   `   
`
vegetable : 3  
`

```python
# write a command that will count the number of fruits and number of vegetables and print a small table
# showing the number of each
fruit_count = 0
vegetable_count = 0
for food in produce_data["produce"]:
    if food.get("type") == "fruit":
        fruit_count = fruit_count + 1
    else:
        vegetable_count = vegetable_count +1

print("{:<9} {:<1} {:<5}".format("fruit", ":", fruit_count))
print("{:<9} {:<1} {:<5}".format("vegetable", ":", vegetable_count))
```

### Exercise 5: 2D Table  (optional)
This is a more advanced exercise.  See if you can figure out an efficient way to accomplish this task.  It is optional for the 3 points for the workbook.  

Now make a 2D table showing counts for both veggie/fruit and count by color.  It should look something like this  
(it is ok to have a different color order)   

|   |green | red | yellow | orange |
|---|:---:|:---:|:---:|:---:|
| fruit | 0 | 2 |  1 |  1 |
| vegetable | 1 | 0 | 1 | 1 |


```python
fruit_count = 0
vegetable_count = 0
fruit_color_counts = {"orange": 0, "yellow": 0, "green": 0, "red": 0}
vegetable_color_counts = {"orange": 0, "yellow": 0, "green": 0, "red": 0}

for food in produce_data["produce"]:
    if food.get("type") == "fruit":
        fruit_count += 1
        color = food.get("color")
        if color in fruit_color_counts:
            fruit_color_counts[color] += 1
    elif food.get("type") == "vegetable":
        vegetable_count += 1
        color = food.get("color")
        if color in vegetable_color_counts:
            vegetable_color_counts[color] += 1

print("{:<9} {:<5} {:<10} {:<11} {:<12} {:<10}".format(" ", "", "orange", "yellow", "green", "red"))
print("{:<9} {:<10} {:<10} {:<10} {:<10} {:<10}".format("fruit", ":", fruit_color_counts["orange"], fruit_color_counts["yellow"], fruit_color_counts["green"], fruit_color_counts["red"]))
print("{:<9} {:<10} {:<10} {:<10} {:<10} {:<10}".format("vegetable", ":", vegetable_color_counts["orange"], vegetable_color_counts["yellow"], vegetable_color_counts["green"], vegetable_color_counts["red"]))
```
