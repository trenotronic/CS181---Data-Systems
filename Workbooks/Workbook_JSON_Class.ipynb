### Reading JSON Files
The json module provides a very easy way to read a json formated file and store it in python.  Notice that "objects" are converted to python dictionaries and "arrays" are converted naturally to python lists.  

```python
# read in a json file 
filename = 'class.json'
file = open(filename,'r')
class_data = json.load(file)
file.close()
print(class_data)
```

```python
# we can also do a pretty print
# the dumps command dumps a json structure to a single string.  
print(json.dumps(class_data,indent=2))
```

### Traversing using the implied structure
Let's traverse the upper level of the json structure and print out all student names as we go.   
This operation implies that we have a fixed json structure (called a schema) and our operation only works with this  
particular format (schema).  

```python
# class_data is a the root and is a python dictionary 
# "class" is the only key in this dictionary
# its value is a list of other dictionaries. 

for student in class_data["class"]:
    
    # each student is now a dictionary with student data inside it
    print(student["student_name"])
    
```

### One Layer Deeper
Let's augment our printing operation to also print all the quiz scores for each student.  



```python
for student in class_data["class"]:
    
    # each student is now a dictionary with student data inside it
    print(student["student_name"])
    for score in student["scores"]:
        if score["type"] == "quiz":
            print("\t",score["name"],":",score["score"])
            
```

```python
def find_matched_values (node,target):
    '''
    This function will start at a json node in a tree and parse the tree from that node
    looking for key/value pairs where the key=target.  It returns a list of all values
    with a key match to the target.  
    This function uses recursion to search down each branch of the tree until it finds a
    matching key/value pair where the key=target string.  It then creates a list of all
    values from those matching keys.  
    '''
    # if this thing is a dictionary (json object)
    ret = []
    if isinstance(node,dict):
        # look for a match at this level, base case
        if target in node:
            # return a list of one value with this match
            ret.append(node[target])
            
        # now we must assume that each value is a potential source for (more) target matches
        # so make a recursive call on each value
        for key in node:
            ret = ret + find_matched_values(node[key],target)
    # now see if this node is a list (json array)
    elif isinstance(node,list):
        for item in node:
            ret = ret + find_matched_values(item,target)

    return ret
```

```python
# useful here to identify individual student names
# which only appear once in the data structure
studentNames = find_matched_values(class_data,'student_name')
print(studentNames)

# Not particularly useful here because there are multiple copies of points
# and our list contains repeated values
# We will have to do a different kind of search for data that is replicated
allPoints = find_matched_values(class_data,'points')
print(allPoints)
```

### Task 1: Extract Scores for an evaluation metric
---
In this task, we will again search the json tree and look for things specific to this schema.  
Namely, we will try to gather all the student scores for a particular evaluation instrument.   

For example, let's assume we want to find all "quiz1" scores.   We will have to search the tree for a value called "scores".  
Once we have this value (a json array, or python list) we can look at each element in the array and see if it matches  
our target name.   

#### Task 1a:  extractScore ( ) function
This function is to take a list of scores and search through this list to see if any of the scores match a particular name.
If there is a match, the function should return the numeric score from this evaluation type, otherwise return None.  

Write this function.   

#### Task 1b:  parseScores ( ) function
This functions job is to recursively search through the json structure to find scores arrays.  If it finds one, it should call the extractScores() function to retrieve the score value and add it to a list.   

- You should assume that the key you are looking for is called "scores".   And the value associated with this key is a list of the student_types of our schema.   
- If you must "hardcode" this search method, that is ok.
- If you can, try to write a general purpose search method that uses recursion to match any key of "scores".  



```python
def extractScore (scoresL, name):
    '''
    This function will take a scores array (a python list) and attempt to find a match
    in that array to the evaluation tool matching the given name.
    PARAMTERS:
    > scoresL is a list representing a score type (list)
    > name is the name attribute of a particular score (string)
    RETURN VALUE:
    We will isolate and return the score associated with the name match or return None if there
    is no match in this score list.  
    '''

    for score in scoresL:
        if score.get("name") == name:
            return score.get("score")
    
    return None


def parseScores (node, name):
    '''
    This function will parse a tree starting at node and look for score types where
    there is a score of name=name.  It will return a list of all such scores.  
    PARAMETERS:
    > node : the current node of the tree (if recursively)
             or the root node (if hardcoding instead of recursion)
    > name : the name of the evaluation_type instrument to match (a string)
    RETURN VALUE:
    > returns a list of numeric scores of all scores in the tree that match the evaluation name
    '''
    
    scores = []

    if isinstance(node, dict):
        for key, value in node.items():
            if key == "scores" and isinstance(value, list):
                score = extractScore(value, name)
                if score is not None:
                    scores.append(score)
            else:
                scores.extend(parseScores(value, name))
                
    elif isinstance(node, list):
        for item in node:
            scores.extend(parseScores(item, name))

    return scores

    
```

#### Testing Task 1

```python
quiz1Grades = parseScores(class_data,'quiz1')
print(quiz1Grades)
# should print [40, 30, 50, None]

midterm1Grades = parseScores(class_data,'mid1')
print(midterm1Grades)
# should print [90, 84, 88, 88]
```

### Task 2: Computing Scores
---
Now let's write two more functions.    The first function will take a list of score types and then compute a student's current grade.   
- The current grade is a weighted percentage of all completed evaluations.
- You will have to iterate through all the evaluations
- Compute the student's percentage on each evaluation
- Weight each student's percentage by the weight value
- Sum a total weight and use that to divide through the accumulated total.

#### Task 2a: computeScore (scoreList)
This function takes a score list and returns a float which is the student's current score level.

#### Task 2b: computeCurrentGrades (node)
This function takes a node (the root at first) and finds all students.  It then calls the computeScore() function for each student to print a table of student name / score.  
It is ok to "hardcode" this function.  You can use the schema structure instead of the recursive search idea.   
When you print the score, use the string format to round it to one decimal place (see the expected output below).  



```python
def computeScore (scoreL):
    '''
    This function computes a students current score in the class.
    PARAMETERS:
    a list of evaluation types (this is the value of the "scores" property)
    RETURN VALUE:
    a float that represents the student's current (weighted) grade in the class
    '''
    total_weighted_score = 0
    total_weight = 0
    
    for score_entry in scoreL:
        score = score_entry.get("score", 0)
        points = score_entry.get("points", 1)  
        weight = score_entry.get("weight", 0)

        # weighted score 
        percentage = (score / points) if points > 0 else 0
        weighted_score = percentage * weight * 100
        total_weighted_score += weighted_score
        total_weight += weight

    # final grade 
    if total_weight > 0:
        return total_weighted_score / total_weight
    return 0.0  
 
def computeCurrentGrades(root):
    '''
    This function iterates through all students and computes a current grade for each one.
    PARAMETERS:
    the root of the json tree for a class object
    RETURN VALUE:
    none, but prints a table of names and grades 
    '''
    if "class" in root:
        for student in root["class"]:
            name = student.get("student_name", "Unknown")
            scores = student.get("scores", [])
            
            current_grade = computeScore(scores)

            print(f"name: {name:<10} score: {current_grade:6.1f}")
```

### Task 3: Adding an evaluation
---
Let's add a final exam score for each student.   Below is a dictionary of final exam grades for each student.  We will assume
- type = "final"
- name = "final"
- points = 200 (max points possible)
- weight = 0.25  (25% of students grade)
- individual scores come from the dictionary below

```python
final_exams = {'Alice':178, 'Bob':190, 'Carol':155, 'Dave':160} 
print(final_exams)
```

#### Task 3b:
This first function once again takes a score list as a parameter.  It adds an evaluation type in the form of a python list with all the appropriate properties (key / value pairs).  
You will call this function for each student.  

#### Task 3c:
The second function takes the root of the tree and the exam dictionary as parameters.  It iterates through the json structure and locates each student.  If the student name is in the exam dictionary, it calls the addEvaluation() function to add a final exam score for that student.  It is ok to hard-code the search process.  

```python
def addEvaluation (scoreL,type,name,points,weight,score):
    '''
    This function adds an evaluation to the score list
    PARAMETERS:
    scoreL : a python list representing an evaluation_type array
    type : python string
    name : python string
    points : python int
    weight : python float
    score : python float (or int) 
    RETURN VALUE:
    none, but the scoreL list is altered by appending a new score to the list
    '''
    evaluation = {
        "type": type,
        "name": name,
        "points": points,
        "weight": weight,
        "score": score
    }

    scoreL.append(evaluation)

def addFinalExams (root,examDict):
    '''
    This function iterates through the class json struct and adds a final exam evaluation for each student.
    PARAMETERS:
    > root: the root node of the json class structure
    > examDict : a dictionary with keys = student names and values = student final exam score
    RETURN VALUE:
    none, but the root structure is altered by adding a new final exam score for each student who has
    a score in the examDict dictionary.  
    '''
    for student in root['class']:
        student_name = student['student_name']
        if student_name in examDict:
            final_score = examDict[student_name]
            addEvaluation(student['scores'], "exam", "final_exam", 200, 0.25, final_score)
```
