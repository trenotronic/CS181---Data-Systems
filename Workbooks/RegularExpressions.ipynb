### Exercise 1: Parse for Capitalized Words
Write a function which takes a string as input and returns a list of all words in the string that are
capitalized (begin with A to Z)   
For this exercise it is ok to assume that a word is composed only of A-Z and a-z (don't worry about words with punctionation like don't).  

```python
def getCapitalWords (s):
    '''
    This function returns a list of capitalized words found in string s.
    PARAMETERS: 
    s - any python string (it is intended this contains regular English text)
    RETURN VALUE
    a list of all words in s which begin with a capital letter
    '''
    list_of_cap_words = re.findall( r'\b[A-Z]+[a-z]*\b' , s)

    return list_of_cap_words    
    
```

```python
# Testing
s = "How about some honey, Pooh Bear?"
print(getCapitalWords(s))

s = "I am A NITWIT."
print(getCapitalWords(s))

s = "no cApitalIzed wOrds in hERE."
print(getCapitalWords(s))

```

### Exercise 2: Double Speak
In this exercise we will parse a string sentence and isolate all words with double consonents.  
Let us assume that all double consonents will be strictly lower case (don't worry about upper case).  
Let us also assume that all words are comprised solely of letters (no punctuation).   

Hint: Read about "backward refereces" in regular expressions; you will probably have to search for something online since they are not mentioned in the textbook.   This is a good instance to apply them although you can probably make this work without a backward reference.  

```python
def doubleConsonents (s):
    '''
    This function returns a list of words in string s that have double consonents.  
    PARAMETERS: string s with words.  We will assume the words are comprised of only letters.
    RETURN VALUE: a list of words in string s that have double consonents.
    '''

    list_of_DC = [myMatch.group() for myMatch in re.finditer( r'\b\w*(\w)\1\w*\b' , s)]

    
    return list_of_DC
```

```python
s = "I am happy too.  Better thee than tattoo committees.  Odd verbiage stammers happily."
print(doubleConsonents(s))
```

### Exercise 3: [HARD CHALLENGE] Gettysburg Address Modernizer
Try to complete this to the best of your ability.   
On June 1 1865 Abraham Lincoln delivered the Gettysburg address  

`Four score and seven years ago our fathers brought forth on this continent, a new nation, conceived in Liberty, and dedicated to `

This is often heralded as one of the preeminent speeches in history (esp US History), but it does contain some out-dated language. In particular, "four score and seven years" is an odd way of speaking that isn't used today.   This kind of counting goes way back into early (and modern) languages that have different counting schemes.   "Mark Twain" is another such example of an odd counting scheme (mark twain means two fathoms deep (aka 12 feet) and was used by riverboat pilots on the Mississippi).   

You are to write a function that looks for "score counts" and translates them into modern numerals.   You will replace sections of embedded strings and return new versions of those strings with the modern numerals.   For example, if the function took as input  

`4 score and 7 years ago our fathers brought forth on this continent, a new nation, conceived in Liberty, and dedicated to `  

it would return   

`87 years ago our fathers brought forth on this continent, a new nation, conceived in Liberty, and dedicated to `   

#### String Matching Types 

You are two match two types of strings
1.  \<num1\> score and \<num2\>
2.  \<num1\> score

where \<num\> is some integer number (comprised of digits 0-9).    Since "score" stands for twenty, you will multiply the first num1 by 20 and then add (if it is present) the second num2 to obtain the modern integer total.  Substitute out the "score" part of the string and substitute in your modern replacement.  Return the new string.  Notice we will not deal with spelled out digits like "four" but instead assume we see "4".   

If you want a hint see example code I included in the cell below with sub and lambda.  Using sub with lambda makes it super easy!  


```python
def score2num (num1,num2):
    '''
    num1, num2 are both strings of digits or are none
    returns num1*20 + num2 as a string
    '''
    if num1:
        num1 = int(num1)
    else:
        num1 = 0
    if num2:
        num2 = int(num2)
    else:
        num2 = 0
    return str(num1*20 + num2)

def scoreConvert(s):
    '''
    This function replaces outdated passages of counting using the word "score" in a string with modern equivalents.  
    PARAMETER:  
    s - any python string which may contain passages of the form
       <num1> score and <num2>     or
       <num1> score
       where num1 and num2 are integer values.
    RETURN VALUE:
    We will replace either of the two patterns with the numeric equivalent ( num1*20 + num2 ) and return a
    new string that is a copy of s with these replacements. 
    '''
    
    return re.sub(r'(\d)\s*score(?:\s*and\s*(\d+))?' , lambda m: score2num(m.group(1), m.group(2)) , s) 
    

```

```python
s = '4 score and 7 years ago our fathers ...'
print(scoreConvert(s))

s = 'Today I am 1 score and 19 years old.  In 1 score weeks I will be 2 score years old.'
print(scoreConvert(s))
```

#### Hint: Using sub with lambda 

The above problem is a substitution problem -- replacing a matched pattern with a different string.  The challenge is that the replacement string depends on information extracted from the match.  So you can't just specify a regular python string.  What you need for the replacement string is a function that returns a string that is computed using matched information.   However, you don't yet have a variable that is the match object.   Here is where lambda comes in -- it gives you access to a match object WITHIN the sub call!  

In the example below, we are replacing all lowercase vowels with upper case vowels.  It is a trivial example, but it illustrates how to use sub with a lambda function.  You can use a more sophisticated version of this same idea to solve the scoreConvert() function in just 1 line of code!!!   The m parameter to the lambda function is the match object from the re.sub() call.  We can use it to access the group (the string).  


```python
s = 'The creepy weirdo!'

pattern = r'[aeiou]'
t = re.sub(pattern,lambda m: m.group().upper(),s)
print(t)
```
