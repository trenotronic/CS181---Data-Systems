### Exercise 1
Write a function which takes a single string as a parameter and returns the percentage of characters in that string which are a digits (0-9).    

Example:  `s = "Jenny I've got your number\nI need to make you mine\n\nJenny don't change your number\n8675309"`
Returns:  `0.0778`



```python
def percentDigits (s):
    '''
    you create comment header
    '''
    total_chars = len(s)

    if total_chars == 0:
        return 0.0
    
    digit_count = sum(char.isdigit() for char in s)
    
    return round(digit_count / total_chars, 4)
```

```python
s = "Jenny I've got your number\nI need to make you mine\n\nJenny don't change your number\n8675309"
print(percentDigits(s))

s = 'no digits here'
print(percentDigits(s))

s = '1111'
print(percentDigits(s))

# what should you do here?  Probably 0 is the right answer.  
s = ''
print(percentDigits(s))

```

### Exercise 2
Write a function which takes a string as input.   The string will only be composed of words and space -- no punctuation or other characters.   The words will always be lower case letters.   
Your function should create a dictionary where the  
key - a word appearing in the string
value - a count of how many times that word appears

```python
def string2Frequency (s):
    '''
    put docstring here
    '''
    words = s.split()
    
    word_dict = {}

    for word in words:
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1

    return word_dict
```

```python
s = "it was the best of times it was the worst of times"
freqDict = string2Frequency(s)
print(freqDict)

```

### Exercise 3
Write a function which takes a string as input.  The string is composed only of words formed with lower case letters and intervening spaces.   There is no punctuation or other characters other than a..z and spaces.   

The function breaks the string up into words and the returns a dictionary where  
key = word in the string  
value = list of indices where that word appears in the string

```python
def string2Indices (s):
    '''
    you put docstring here
    '''
    words = s.split()

    word_dict = {}

    for index, word in enumerate(words):
        if word in word_dict:
            word_dict[word].append(index)
        else:
            word_dict[word] = [index]

    return word_dict
```

```python
s = "it was the best of times it was the worst of times"
idxDict = string2Indices(s)
print(idxDict)
```

### Exercise 4
Write a function which takes a DNA string as input and returns the DNA string in sorted alphabetical order.  
Recall that all DNA strings contain  A, C, G, T values.   Those are the only four characters that will appear in the string.   

Hint: because there are only four characters, think about how to count the characters and use that count to generate a new string.   This will be much more efficient than implementing a regular sorting routine like insertionSort.  
Hint: use a dictionary to keep track of the count!  

```python
def sortDNA (dnaString):
    '''
    put a docstring here
    '''
    count = { 'A' : 0 , 'C' : 0 , 'G' : 0 , 'T' : 0 }

    for char in dnaString:
        if char in count:
            count[char] += 1
        else:
            raise ValueError(f"Invalid character '{char}' found in DNA string")

    sorted_dna = ( 'A' * count['A'] + 'C' * count['C'] + 'G' * count['G'] + 'T' * count['T'] )
    
    return sorted_dna
        
```

```python
s = 'AGCTACCATGA'
print(sortDNA(s))
s = 'ACAC'
print(sortDNA(s))
```

### Exercise 5
Write a function that takes a string s as input and returns a list of all possible unique substrings.    

Super Guru Suggestion:   If you use a list, you will spend O(n) time each time you test if something is in that list because python has to do a linear search.  Use a dictionary instead to keep track of unique substrings.  Then extract a list of keys from the dictionary.  

```python
def allSubstrings(s):
    '''
    you write docstring
    '''
    pos_substrings = {}
               
    for i in range(len(s)):
        for j in range(i + 1, len(s) + 1):
               substr = s[i:j]
               pos_substrings[substr] = True

    return list(pos_substrings.keys())

        
```

```python
s = 'aaba'
print(allSubstrings(s))
```

### Exercise 6
Write a function that takes a list of strings as input.  This function is to return a new list which is a set of all possible strings that can be formed by taking two strings from the input list and appending them together.  

Note, if we combine this with the "all substrings" function, then we are effectively solving the problem:  
Generate all possible strings which can be formed by concatenating two substrings of s.  

Hint: remember the product command!  

```python
from itertools import product

def allStringPairs (s):
    '''
    This function takes a list s of strings.  It returns a new list which is comprised of all possible unique strings
    that can be formed by taking two strings from s.  
    '''
    combined_strings = {a + b for a, b in product(s, repeat=2)}
    return list(combined_strings)
    
```

```python
s = 'aaba'
substrings = allSubstrings(s)
allPairsOfSubstrings = allStringPairs(substrings)
print(allPairsOfSubstrings)
```
