### Exercise 3.3  Binary Vector Multiplication
Write a binary operation to perform vector multiplication.  Hint: use the zip function.  

```python
def vectorMultiply (vec1,vec2):
    '''
    vectorMultiply(vec1,vec2)
    This function returns a new list with the elements of vec1 and vec2 multiplied 
    together to create the return list.
    Parameters:
        vec1: a list of n integer or float items
        vec2: a list of m integer or float items
        Note: if the lists are not the same length, the return list will be the length
        of the shorter of the two input lists.
    Return Value:
        a vector of length n or m (smaller of the two) containing element wise 
        multiplication of the input vectors. 
    '''
    new_vec = []
    for a ,b in zip(vec1, vec2):
        new_vec.append( a * b ) 
    return new_vec
```

```python
# Test 1
v1  = [1,2,3]
v2 = [4,5,6]
print(vectorMultiply(v1,v2))

# Test 2
v1  = [1,2,3]
v2 = [4,5,6,7]
print(vectorMultiply(v1,v2))

```

### Exercise: Finding locations of capital letters
Write a python function that returns a list of all indices in a string that contain a capital letter.   
Hint: use enumerate   
Hint: look at the string constant ascii_uppercase


```python
def locateCapitalLetters (s):
    '''
    This function searches string s to find the indices of all capital letters.
    PARAMETER:
    s - any string
    RETURN VALUE:
    a list containing the indices where a capital letter appears in s.
    '''
    indices = []  # initialize an empty list

    for index, char in enumerate(s):  # loop through the string
        if char in string.ascii_uppercase:  # check for capital letters
            indices.append(index)  

    return indices  
```

```python
# Test 1
s1 = 'What is the Air Speed Velocity of an English Sparrow?'
print(locateCapitalLetters(s1))

# Test 2
s2 = 'this string has no capital letters.'
print(locateCapitalLetters(s2))

# Test 3
s3 = '#$%_++~'
print(locateCapitalLetters(s3))
```

### Exercise: Creating a deck of playing cards and then drawing a poker hand.
In this exercise we will create a deck of 52 playing cards and then return a five card poker hand as a list.   
Hint: use product  


```python
def createDeck ():
    '''
    This function creates a list of 52 strings.  The strings are the playing cards in a deck.
    PARAMETER: none
    RETURN VALUE: a list of 52 strings ['Ac','1c','2c',...,'Qs','Ks'] representing playing cards in a deck
    '''
    # nums is a list of playing card denominations as a list of strings
    nums = ['A']
    for i in range(2,11):
        nums.append(str(i))
    nums.append('J')
    nums.append('Q')
    nums.append('K')

    # suits is a list of card suits as a list of strings
    suits = ['c','d','h','s']     #clubs, diamonds, hearts, spades

    return [ (nums + suits) for ( nums, suits) in itertools.product (nums, suits) ]


def drawHand (deck):
    '''
    This function returns five randomly selected items from the deck of cards for a poker hand.
    PARAMETER: a list of playing cards (possibly 52 of them in a deck)
    RETURN VALUE: a list of 5 of these strings randomly selected
    *** The deck changes composition because the five cards are removed from the deck ***
    '''

    random_cards = random.sample(deck, 5)
    
    # Remove the selected cards from the deck
    for card in random_cards:
        deck.remove(card)
    
    return random_cards

```

```python
# Test 1  create deck of cards
deck = createDeck()
print(deck)

#Test 2  create some random poker hands
for i in range(4):
    hand = drawHand(deck)
    print(hand)

print('Remaining deck: ',deck)
```
